```python
from galpy.potential import NFWPotential, HernquistPotential
from galpy.potential import plotRotcurve
from galpy.orbit import Orbit
```

```python
np= NFWPotential(a=3.5,normalize=.35)
o = Orbit(vxvv=[2.0,-0.5,1.0,0.1,0.0,pi/2.0]) #[R,vR,vT(,z,vz,phi)]
ts = linspace(0,200,10000)
o.integrate(ts,np,method='odeint')
```

```python
o.plot(d1='x',d2='y')
```
